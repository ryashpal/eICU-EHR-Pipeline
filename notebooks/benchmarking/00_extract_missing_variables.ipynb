## Read necessary files

```python
import os
import pandas as pd


labsDf = pd.read_csv(os.environ['EICU_EHR_PIPELINE_BASE'] + '/data/eICU/lab.csv')
datamatrixDf = pd.read_csv(os.environ['EICU_EHR_PIPELINE_BASE'] + '/data/final/data_matrix.csv')

labsDf.shape, datamatrixDf.shape
```

```python
apachePredDf = pd.read_csv('/superbugai-data/physionet.org/physionet.org/files/eicu-crd/2.0/apachePredVar.csv')
apacheApsDf = pd.read_csv('/superbugai-data/physionet.org/physionet.org/files/eicu-crd/2.0/apacheApsVar.csv')
admissionDxDf = pd.read_csv('/superbugai-data/physionet.org/physionet.org/files/eicu-crd/2.0/admissionDx.csv')

apachePredDf.shape, apacheApsDf.shape, admissionDxDf.shape
```

```python
import os
import pandas as pd


patientsDf = pd.read_csv(os.environ['EICU_EHR_PIPELINE_BASE'] + '/data/eICU/patient.csv')

patientsDf.shape
```

```python
datamatrixDf
```

## Obtain `potassium` results for the cohort

```python
potassiumDf = labsDf[(labsDf.labname == 'potassium') & (labsDf.patientunitstayid.isin(datamatrixDf.visit_occurrence_id))]
potassiumDf
```

```python
potassiumDf = potassiumDf.merge(patientsDf[['patientunitstayid', 'unitdischargeoffset']], on='patientunitstayid')
potassiumDf
```

```python
potassiumDf = potassiumDf[(potassiumDf.labresultoffset > potassiumDf.unitdischargeoffset - (60*24)) & (potassiumDf.labresultoffset < potassiumDf.unitdischargeoffset)]
potassiumDf
```

```python
potassiumDf['sortcolumn'] = (potassiumDf.labresult - 4).abs()
potassiumDf
```

```python
potassiumDf = potassiumDf.sort_values(['patientunitstayid','sortcolumn'], ascending=False).groupby('patientunitstayid').first().reset_index()
potassiumDf
```

```python
potassiumDf[['patientunitstayid', 'labresult']].to_csv(os.environ['EICU_EHR_PIPELINE_BASE'] + '/data/saps_ii/potassium.csv', index=False)
```

## Obtain `bicarbonate` results for the cohort

```python
bicarbonateDf = labsDf[(labsDf.labname == 'bicarbonate') & (labsDf.patientunitstayid.isin(datamatrixDf.visit_occurrence_id))]
bicarbonateDf
```

```python
bicarbonateDf = bicarbonateDf.merge(patientsDf[['patientunitstayid', 'unitdischargeoffset']], on='patientunitstayid')
bicarbonateDf
```

```python
bicarbonateDf = bicarbonateDf[(bicarbonateDf.labresultoffset > bicarbonateDf.unitdischargeoffset - (60*24)) & (bicarbonateDf.labresultoffset < bicarbonateDf.unitdischargeoffset)]
bicarbonateDf
```

```python
bicarbonateDf = bicarbonateDf.sort_values(['patientunitstayid', 'labresult'], ascending=True).groupby('patientunitstayid').first().reset_index()
bicarbonateDf
```

```python
bicarbonateDf[['patientunitstayid', 'labresult']].to_csv(os.environ['EICU_EHR_PIPELINE_BASE'] + '/data/saps_ii/bicarbonate.csv', index=False)
```

## Obtain `Type of admission` for the cohort

```python
patientsDf
```

```python
dataDf = datamatrixDf[['visit_occurrence_id']]\
    .merge(admissionDxDf[['patientunitstayid', 'admitdxpath']], left_on='visit_occurrence_id', right_on='patientunitstayid')\
    .drop(columns=['patientunitstayid'])\
    .merge(patientsDf[['patientunitstayid', 'hospitaladmitsource', 'age']], left_on='visit_occurrence_id', right_on='patientunitstayid')\
    .drop(columns=['patientunitstayid'])\
    .merge(apachePredDf[['patientunitstayid', 'electivesurgery']], left_on='visit_occurrence_id', right_on='patientunitstayid')\
    .drop(columns=['patientunitstayid'])
dataDf
```

```python
dataDf['surgicalpatient'] = (dataDf.admitdxpath.str.contains('All Diagnosis|Operative') & dataDf.hospitaladmitsource.isin(['Operating Room', 'Recovery Room', 'PACU']))
dataDf
```

```python
dataDf['Scheduled surgical'] = ((dataDf.surgicalpatient) & (dataDf.electivesurgery == 1))
dataDf
```

```python
dataDf['Unscheduled surgical'] = ((dataDf.surgicalpatient) & (~dataDf['Scheduled surgical']))
dataDf
```

```python
dataDf['Medical'] = ((~dataDf['Scheduled surgical']) & (~dataDf['Unscheduled surgical']))
dataDf
```

```python
dataDf['admissiontype'] = dataDf.apply(lambda x: ('Scheduled surgical' if x['Scheduled surgical'] else ('Unscheduled surgical' if x['Unscheduled surgical'] else 'Medical')) , axis = 1)
dataDf
```

```python
dataDf.to_csv(os.environ['EICU_EHR_PIPELINE_BASE'] + '/data/saps_ii/admissiontype.csv', index=False)
```
