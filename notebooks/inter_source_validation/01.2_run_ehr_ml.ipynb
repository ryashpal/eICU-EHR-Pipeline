# Run EHR-ML

## Evaluate mortality prediction for eICU data

```python
import os


for (wb, wa) in [(30, 30)]:
    print('--------------------------------------------------------------------------------------------------------------------------------------------------------------------')
    print('---------------------------------------------------------- wb: ' + str(wb) + ' --- wa: ' + str(wa) + ' -------------------------------------------------------------')
    print('--------------------------------------------------------------------------------------------------------------------------------------------------------------------')
    os.system(
        'cd ' + os.environ['EHR_ML_BASE'] + ';'
        +
        '''.venv/bin/python -m ehrml.ensemble.Evaluate ''' + os.environ['EICU_EHR_PIPELINE_BASE'] + '''/data/experiments/inter_source_validation/data/data_matrix_eicu_standardised.csv -tc "death_0_30_adm" -ic "person_id" "visit_occurrence_id" -mdc "measurement_date" -adc "visit_start_date_adm" -wb ''' + str(wb) + ''' -wa ''' + str(wa) + ''' -sp ''' + os.environ['EICU_EHR_PIPELINE_BASE'] + '''/data/experiments/inter_source_validation/evaluation/eicu_wb_''' + str(wb) + '''_wa_''' + str(wa) + '''_standardised.json;'''
        )
```

## Performs predictions for MIMIC using model built on eICU data
