# Calculating computational time

## Computational time calculation for time-window analysis

```python
import os


with open(os.environ['EICU_EHR_PIPELINE_BASE'] + '/data/experiments/01_time_window_analysis/ehr-ml.log', 'r') as f:
    logs = f.read()
```

```python
logLines = logs.split('\n')
```

```python
filteredLogLines = [line for line in logLines if ('INFO - Starting - windowStart' in line) or ('INFO - Saving results' in line)]
```

```python
combinedLogLines = []
for i in range(int(len(filteredLogLines)/2)):
    combinedLogLines.append(filteredLogLines[(i * 2) + 1] + ' ' + filteredLogLines[(i * 2)])
```

```python
combinedLogLines
```

```python
import pandas as pd

logsDf = pd.DataFrame([(line[:23], line[57:80], line[122:123], line[134:]) for line in combinedLogLines], columns=['end_time', 'start_time', 'window_start', 'window_end'])
logsDf['end_time'] = pd.to_datetime(logsDf['end_time'], format='%Y-%m-%d %H:%M:%S,%f')
logsDf['start_time'] = pd.to_datetime(logsDf['start_time'], format='%Y-%m-%d %H:%M:%S,%f')
logsDf['window_start'] = logsDf['window_start'].astype(int)
logsDf['window_end'] = logsDf['window_end'].astype(int)
logsDf['time_taken'] = round((logsDf.end_time - logsDf.start_time).dt.total_seconds()/60, 2)
```

```python
logsDf.sort_values(by=['window_start', 'window_end'])[['window_start', 'window_end', 'time_taken']]
```
